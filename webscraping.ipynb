```python
import requests
from bs4 import BeautifulSoup
import requests
# import gspread
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
# from gspread_dataframe import set_with_dataframe
import requests
import selenium.webdriver as webdriver
import time
import json
import numpy as np
from datetime import datetime
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.firefox.options import Options


firefox_binary_path = r"C:\Program Files\Mozilla Firefox\firefox.exe"

# Create an instance of the Firefox Options class
options = Options()

# Set the binary location in the options
options.binary_location = firefox_binary_path



url="https://www.magicbricks.com/property-for-rent/residential-real-estate?proptype=Multistorey-Apartment,Builder-Floor,Studio-Apartment,Penthouse,Residential-House,Villa&cityName=Bangalore&possessionYears=availfrom1&pageOption=A"
page=requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

housetitle_list=soup.find_all('div',class_='mb-srp__list')

driver = webdriver.Firefox()

driver.get(url)

# soup = BeautifulSoup(driver.page_source, 'html.parser')

house_list=[]
for _ in range(50):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    houses = soup.find_all('div',{'class':'mb-srp__list'})
    for i in range(len(houses)):
        card=houses[i]
        housetitle=card.find('h2',{'class':'mb-srp__card--title'}).text
        builder_=card.find('a',{'class':'mb-srp__card__society--name'})
        if builder_:
            builder=builder_.text
        else:
            builder=np.nan
    
        feature_main=card.find_all('div',{'class':'mb-srp__card__summary__list--item'}) 
        furnishing=feature_main[0].find('div',{'mb-srp__card__summary--value'}).text
        bathroom=feature_main[1].find('div',{'class':'mb-srp__card__summary--value'}).text
        tenant_prefered=feature_main[2].find('div',{'class':'mb-srp__card__summary--value'}).text
        availability=feature_main[3].find('div',{'class':'mb-srp__card__summary--value'}).text
        area=feature_main[4].find('div',{'class':'mb-srp__card__summary--value'}).text
        floor=feature_main[5].find('div',{'class':'mb-srp__card__summary--value'}).text
    
        pf_=card.find('div',{'class':'mb-srp__card__usp'})
        if pf_:
            property_facing=pf_.find('div',{'class':'mb-srp__card__usp--item'}).text
        else:
            property_facing=np.nan
        notes=card.find('div',{'class':'mb-srp__card--desc--text'}).text
        rent=card.find('div',{'class':'mb-srp__card__price--amount'}).text
        ul_tag = soup.find('ul', class_='mb-srp__card__kilometre__list')
        if ul_tag:
            li_tags = ul_tag.find_all('li')
            nearby=', '.join(item.text for item in li_tags)
        script_tags = soup.find_all('script', type="application/ld+json")
        link = script_tags
        if len(script_tags) >= 1:
            # Extract URL from the second script tag
            json_data = script_tags[1].string
            if json_data:
                json_data = json.loads(json_data)
                if 'url' in json_data:
                    url = json_data['url']
                    link=url
        else:
            link=np.nan
        data= {
            'HouseTitle' : housetitle,
            'Builder': builder,
            'Furnishing':furnishing,
            'Bathroom':bathroom,
            'TenantPrefered':tenant_prefered,
            'Availability':availability,
            'Area':area,
            'Floor':floor,
            'PropertyFacing':property_facing,
            'Notes':notes,
            'Rent':rent,
            'NearBy':nearby,
            'Link':link
    
        }
        house_list.append(data)

df=pd.DataFrame(house_list)
# # print(df)
df.to_csv('Rent_houses_dataset1.csv',index=False)
driver.quit()
```
